```python
import numpy as np
import matplotlib.pyplot as plt
import secrets
from matplotlib.pyplot import figure
import math

# importing required libraries
```

```python
def f(x):
    return np.tan(x)-x
    #return x-np.cos(x)
    #return x-np.sin(x)
    #return x**3-2*x+5
    #return 816*(x**3)-3835*(x**2)+6000*x-3125

# defining the function
```

```python
def df(x):
    return np.tan(x)*np.tan(x)
    #return 1+np.sin(x)
    #return 1-np.cos(x)
    #return 3*(x**2)-2
    #return 2448*(x**2)-7670*x+6000
    
# defining the derivative of the function
```

```python
np.set_printoptions(suppress=True)

# preventing printing the numbers with scientific notation
```

```python
x_min = -10
x_max = 10
y_min = -10
y_max = 10
print("x_min =", x_min)
print("y_max =", x_max)
print("y_min =", y_min)
print("y_max =", y_max)
```

```python
step = 0.1
stepinv = int(1/step)
print("step =", step)
```

```python
x = np.arange(x_min, x_max, step)
```

```python
n = 100
print("n =", n)

# n is number of random values to be generated
```

```python
vals = x_min + np.random.rand(1,n)*(x_max-x_min)        # generating random values in the given interval
pos = []                                                # stores values, where f(x) > 0
neg = []                                                # stores values, where f(x) < 0
seeds = []                                              # stores potential initial values for root-finding algorithms
```

```python
vals
```

```python
f(vals)
```

```python
for i in range(0,n):
    if(f(vals[0][i]) > 0):
        pos.append(vals[0][i])
    else:
        neg.append(vals[0][i])
```

```python
pos
```

```python
neg
```

```python
for i in range (0,len(pos)):
    for j in range(0,len(neg)):
        seeds.append((neg[j]*f(pos[i])-f(neg[j])*pos[i])/(f(pos[i]-f(neg[j]))))

# using interpolation method to arrive at a seed for a root-finding algorithm
```

```python
seeds
```

```python
max_iters = 20
print("max_iters =", max_iters)
```

```python
rows, cols = (len(seeds), max_iters)
nriters = [[0 for i in range(cols)] for j in range(rows)]

# storing each seed and it's convergence after each iteration for Newton-Raphson method
```

```python
for i in range(0,len(seeds)):
    nriters[i][0] = seeds[i]
    
# starting value of Newton-Raphson method is the seed
```

```python
for i in range(0,len(seeds)):
    for j in range(0,max_iters-1):
        nriters[i][j+1] = nriters[i][j] - (f(nriters[i][j])/df(nriters[i][j]))

# applying Newton-Raphson method for each possible seed
```

```python
for i in range(0,len(seeds)):
    print(nriters[i][max_iters-1])
```

```python
x_step = 1
y_step = 1
tolerance = 1
print("x_step =", x_step)
print("y_step =", y_step)
print("tolerance for seeds =", tolerance)
```

```python
for i in range(0,len(seeds)):
    if(abs(f(seeds[i])) < tolerance):
        print(seeds[i])
```

```python
figure(figsize=(20, 20), dpi=100)
plt.xticks(np.arange(x_min, x_max, x_step))
plt.yticks(np.arange(y_min, y_max, y_step))
plt.plot(x, f(x))        # plot the function
plt.axis([x_min,x_max,y_min,y_max])
plt.plot(x, np.zeros(stepinv*(x_max-x_min)), color = 'black')        # X-axis
plt.plot(np.zeros(stepinv*(x_max-x_min)), x, color = 'black')        # Y-axis
plt.plot(x, np.full(stepinv*(x_max-x_min), tolerance), color = 'yellow')         # upper bound for tolerance
plt.plot(x, np.full(stepinv*(x_max-x_min), -tolerance), color = 'yellow')        # lower bound for tolerance
for i in range(0,len(seeds)):
    if(abs(f(seeds[i])) < tolerance):
        plt.plot(seeds[i],f(seeds[i]), 'gx', color = 'red', markersize = 5)
        plt.plot(nriters[i][max_iters-1],f(nriters[i][max_iters-1]), 'go', color = 'green', markersize = 10)
        plt.annotate(round(nriters[i][max_iters-1], 2), (nriters[i][max_iters-1], f(nriters[i][max_iters-1])), fontsize = 20)
    else:
        plt.plot(seeds[i],f(seeds[i]), 'gx', color = 'purple', markersize = 5)
        
# blue is the function plotted
# red crosses are the useful seeds generated
# purple crosses are the useless seeds generated
# green dots are the approximate zeroes of the function
# yellow lines are the tolerance bounds
```
